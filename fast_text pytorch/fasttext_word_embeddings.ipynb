```python
from gensim.models import FastText
from nltk.corpus import brown
from gensim.models import Word2Vec
from gensim.models.word2vec import Text8Corpus

ft_model = FastText(Text8Corpus("/content/text8"), size=100, window=5, min_count=150, workers=4, min_n=3, max_n =10)
```

```python
import nltk
nltk.download('brown')

print(brown.sents())
w2v_model = Word2Vec(brown.sents(), size=128, window=5, min_count=3, workers=4)
```

```python
# get most similar words
print(ft_model.wv.most_similar('paris'))
```

```python
# "King" - "Man" + "Woman" == "Queen"
print(ft_model.most_similar(['woman', 'king'], ['man'], topn=3))
```

```python
# "Father" - "Boy" + "Girl" == "Mother"
print(ft_model.most_similar(['girl', 'father'],['boy'], topn=3))
```

```python
# "Father" - "Boy" + "Girl" == "Mother"
print(ft_model.most_similar(['paris', 'italy'],['france'], topn=3))
```

```python
# misspell something similar to Venice and we still get a vector ...
print(ft_model.wv['veniciaaaaaa'])
```

```python
# let us see if it still understood wewere saying venice
print(ft_model.most_similar('veniciaaaaaa', topn=3))
```

# Visualization
fastText vectors are still numpy vectors. We can use the same code to create the word cloud. 

```python
x = []
for word in ft_model.wv.vocab:
    x.append(ft_model.wv[word])
```

```python
import numpy as np 
import pandas as pd 
from sklearn.manifold import TSNE 
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure 
from bokeh.models import ColumnDataSource, LabelSet
output_notebook()
```

```python
x = np.array(x)
print('computed x: ', x.shape)
x_embeeded = TSNE(n_components=2, n_iter=250, verbose=2).fit_transform(x)
print("Computed t-SNE", x_embeeded.shape)
```

```python
w2v_model2 = Word2Vec(Text8Corpus('/content/text8'), size=100, window=5, min_count=150, workers=4)
```

```python
df = pd.DataFrame(columns=['x', 'y', 'word'])
df['x'], df['y'], df['word'] = x_embeeded[:,0], x_embeeded[:,1], w2v_model2.wv.vocab

source = ColumnDataSource(ColumnDataSource.from_df(df))
labels = LabelSet(x='x', y='y', text='word', y_offset=8,
                  text_font_size='8pt', text_color='#555555',
                  source=source, text_align='center')

plot = figure(plot_width=600, plot_height=600)
plot.circle('x', 'y', size=12, source=source, line_color='black', fill_alpha=0.8)
plot.add_layout(labels)
show(plot, notebook_handle=True)
```
